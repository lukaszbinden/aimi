```python
# Define constants
HOUNSFIELD_MIN = -1000
HOUNSFIELD_MAX = 2000
HOUNSFIELD_RANGE = HOUNSFIELD_MAX - HOUNSFIELD_MIN

SLICE_X = True
SLICE_Y = True
SLICE_Z = False

IMAGE_HEIGHT = 320
IMAGE_WIDTH_ORIG = 15
IMAGE_WIDTH = 16
IMG_SIZE = (IMAGE_HEIGHT, IMAGE_WIDTH)
```

```python
def normalizeImageIntensityRange(img):
    # img[img < HOUNSFIELD_MIN] = HOUNSFIELD_MIN
    # img[img > HOUNSFIELD_MAX] = HOUNSFIELD_MAX
    # return (img - HOUNSFIELD_MIN)/HOUNSFIELD_RANGE
    return (img - np.min(img)) / np.max(img)
```

```python
targetName = 'prostate_31'
base = "/home/lukas/datasets/prostate/Task05_Prostate/"
targetImagePath = f'{base}/imagesTr/{targetName}.nii.gz'
targetMaskPath  = f'{base}/labelsTr/{targetName}.nii.gz'

imgTargetNii = nib.load(targetImagePath)
imgMaskNii = nib.load(targetMaskPath)

imgTargetNii_fdata = imgTargetNii.get_fdata()
imgTargetNii_fdata = imgTargetNii_fdata[:,:,:,0]
np.min(imgTargetNii_fdata), np.max(imgTargetNii_fdata), imgTargetNii_fdata.shape, type(imgTargetNii_fdata)

```

```python
imgTarget = normalizeImageIntensityRange(imgTargetNii_fdata)
imgMask = imgMaskNii.get_fdata()

np.min(imgTarget), np.max(imgTarget), imgTarget.shape, type(imgTarget)
```

```python
model = load_model('UNET-ProstateSegmentation_320_16.h5')
```

### Single slicing prediction

```python
def scaleImg(img, height, width):
    return cv2.resize(img, dsize=(width, height), interpolation=cv2.INTER_LINEAR)
```

### Predicting a full volume

```python
def predictVolume(inImg, toBin=True):
    (xMax, yMax, zMax) = inImg.shape
    
    outImgX = np.zeros((xMax, yMax, zMax))
    outImgY = np.zeros((xMax, yMax, zMax))
    outImgZ = np.zeros((xMax, yMax, zMax))
    
    cnt = 0.0
    if SLICE_X:
        cnt += 1.0
        for i in range(xMax):
            img = scaleImg(inImg[i,:,:], IMAGE_HEIGHT, IMAGE_WIDTH)[np.newaxis,:,:,np.newaxis]
            tmp = model.predict(img)[0,:,:,0]
            outImgX[i,:,:] = scaleImg(tmp, yMax, zMax)
    if SLICE_Y:
        cnt += 1.0
        for i in range(yMax):
            img = scaleImg(inImg[:,i,:], IMAGE_HEIGHT, IMAGE_WIDTH)[np.newaxis,:,:,np.newaxis]
            tmp = model.predict(img)[0,:,:,0]
            outImgY[:,i,:] = scaleImg(tmp, xMax, zMax)
    if SLICE_Z:
        cnt += 1.0
        for i in range(zMax):
            img = scaleImg(inImg[:,:,i], IMAGE_HEIGHT, IMAGE_WIDTH)[np.newaxis,:,:,np.newaxis]
            tmp = model.predict(img)[0,:,:,0]
            outImgZ[:,:,i] = scaleImg(tmp, xMax, yMax)
        
    # TODO this line is not clear to me
    outImg = (outImgX + outImgY + outImgZ)/cnt
    if toBin:
        outImg[outImg>0.5] = 1.0
        outImg[outImg<=0.5] = 0.0
    return outImg
```

```python
predImg = predictVolume(imgTarget)
```

### Convert binary image to mesh

```python
from skimage.measure import marching_cubes_lewiner
```

```python
vertices,faces,_,_ = marching_cubes_lewiner(predImg)
```

### Save mesh to file (.stl)

```python
from stl import mesh
```

```python
def dataToMesh(vert, faces):
    mm = mesh.Mesh(np.zeros(faces.shape[0], dtype=mesh.Mesh.dtype))
    for i, f in enumerate(faces):
        for j in range(3):
            mm.vectors[i][j] = vert[f[j],:]
    return mm
```

```python
mm = dataToMesh(vertices, faces)
mm.save('prostate-segmented.stl')
```
