```python
# CONSTANTS!!!

# STEP 1 - Load and visualize data
dataInputPath = '/home/lukas/sample_data/volumes/'
dataInputPath = '/home/lukas/datasets/prostate/Task05_Prostate'
imagePathInput = os.path.join(dataInputPath, 'imagesTr/')
maskPathInput = os.path.join(dataInputPath, 'labelsTr/')

dataOutputPath = '/home/lukas/datasets/prostate/Task05_Prostate/slices_dennis'
imageSliceOutput = os.path.join(dataOutputPath, 'img/')
maskSliceOutput = os.path.join(dataOutputPath, 'mask/')

# STEP 2 - Image normalization
HOUNSFIELD_MIN = -1000
HOUNSFIELD_MAX = 2000
HOUNSFIELD_RANGE = HOUNSFIELD_MAX - HOUNSFIELD_MIN

# STEP 3 - Slicing and saving
SLICE_X = True
SLICE_Y = True
SLICE_Z = True

SLICE_DECIMATE_IDENTIFIER = 3
```

```python
 # Load image and see max min Hounsfield units
imgPath = os.path.join(imagePathInput, 'prostate_32.nii.gz')
img = nib.load(imgPath).get_fdata()
img = img[:,:,:,0]
np.min(img), np.max(img), img.shape, type(img)
```

```python
# Load image mask and see max min Hounsfield units
maskPath = os.path.join(maskPathInput, 'prostate_32.nii.gz')
mask = nib.load(maskPath).get_fdata()
np.min(mask), np.max(mask), mask.shape, type(mask)
```

```python
# Show image slice
imgSlice = img[:,:,1]
plt.imshow(imgSlice, cmap='gray')
plt.show()
```

```python
# Show image slice
imgSlice = mask[:,:,10]
plt.imshow(imgSlice, cmap='gray')
plt.show()
```

```python
# Normalize image
def normalizeImageIntensityRange(img):
    return (img - np.min(img)) / np.max(img)

nImg = normalizeImageIntensityRange(img)
np.min(nImg), np.max(nImg), nImg.shape, type(nImg)
```

```python
# Read image or mask volume
def readImageVolume(imgPath, normalize=False):
    img = nib.load(imgPath).get_fdata()
    if len(img.shape) == 4:
        img = img[:,:,:,0]
    if normalize:
        return normalizeImageIntensityRange(img)
    else:
        return img
    
readImageVolume(imgPath, normalize=True)
readImageVolume(maskPath, normalize=False)
```

```python
# Save volume slice to file
def saveSlice(img, fname, path):
    img = np.uint8(img * 255)
    fout = os.path.join(path, f'{fname}.png')
    cv2.imwrite(fout, img)
    # print(f'[+] Slice saved: {fout}', end='\r')
```

```python
saveSlice(nImg[:,:,10], 'test', imageSliceOutput)
saveSlice(mask[:,:,10], 'test', maskSliceOutput)
```

```python
# Slice image in all directions and save
def sliceAndSaveVolumeImage(vol, fname, path):
    (dimx, dimy, dimz) = vol.shape
    print(dimx, dimy, dimz)
    cnt = 0
    if SLICE_X:
        cnt += dimx
        print('Slicing X: ')
        for i in range(dimx):
            saveSlice(vol[i,:,:], fname+f'-slice{str(i).zfill(SLICE_DECIMATE_IDENTIFIER)}_x', path)
            
    if SLICE_Y:
        cnt += dimy
        print('Slicing Y: ')
        for i in range(dimy):
            saveSlice(vol[:,i,:], fname+f'-slice{str(i).zfill(SLICE_DECIMATE_IDENTIFIER)}_y', path)
            
    if SLICE_Z:
        cnt += dimz
        print('Slicing Z: ')
        for i in range(dimz):
            saveSlice(vol[:,:,i], fname+f'-slice{str(i).zfill(SLICE_DECIMATE_IDENTIFIER)}_z', path)
    return cnt
```

```python
# Read and process image volumes
for index, filename in enumerate(sorted(glob.iglob(imagePathInput+'*.nii.gz'))):
    img = readImageVolume(filename, True)
    print(index, filename, img.shape, np.sum(img.shape), np.min(img), np.max(img))
    fnew = 'prostate'+str(index)
    numOfSlices = sliceAndSaveVolumeImage(img, fnew, imageSliceOutput)
    print(f'\n{filename} -> {fnew}, {numOfSlices} slices created \n')
```

```python
# Read and process image mask volumes
for index, filename in enumerate(sorted(glob.iglob(maskPathInput+'*.nii.gz'))):
    img = readImageVolume(filename, False)
    print(filename, img.shape, np.sum(img.shape), np.min(img), np.max(img))
    fnew = 'prostate'+str(index)
    numOfSlices = sliceAndSaveVolumeImage(img, fnew, maskSliceOutput)
    print(f'\n{filename} -> {fnew}, {numOfSlices} slices created \n')
```
